```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel
```

```python
A = np.array([ 82, 72, 87, 95, 86, 95, 98, 74, 75, 82, 85, 86, 98, 99, 57,
              83, 90, 100, 97, 82, 96, 96, 86, 100, 90, 62, 51, 84, 52, 
              100, 100, 65, 82])

B = np.array([ 76, 63, 100, 100, 67, 88, 100, 82, 90, 65, 74, 62, 77, 100,
              100, 89, 83, 56, 100, 99, 89, 88, 84, 72, 71, 60, 62, 82,
              66, 72, 62, 49, 95])
```

```python
t_stat, p_value = ttest_rel(A, B, alternative = 'less')

print("Test statistic is {}".format(np.round(t_stat,4)))
print("p-value is {}".format(np.round(p_value, 4)))

a = 0.05

if p_value > a:
    print("유의수준 0.05에서 기말고사 점수가 중간고사 보다 올랐습니다.")
else:
    print("유의수준 0.05에서 기말고사 점수가 중간고사 보다 오르지 않았습니다..")
    
```
