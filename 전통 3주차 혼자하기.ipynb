```python
n = 10
s = np.arange(1,501)
s_mean = np.zeros(n)

for i in range(n):
    x = np.random.choice(s, 5, replace = False)
    s_mean[i] = np.mean(x)

plt.hist(s_mean)
plt.title("Sample means : n=10")
```

```python
from scipy.stats import f
dfn = 40
dfd = 8

x = np.linspace(f.ppf(0.01, dfn, dfd), f.ppf(0.99, dfn, dfd), 100)
fv = f(dfn, dfd)

fig, ax = plt.subplots(1,1)
ax.plot(x, fv.pdf(x), 'g-', lw=1)
plt.title("F distribution: F(40,8)")
```

```python
#  Python SciPy  Matplotlib  F      .     .

#1. `from scipy.stats import f`: SciPy `stats`  F      .

#2. `dfn = 40`  `dfd = 8`:  (`dfn`)  (`dfd`) .  F    .

#3. `x = np.linspace(f.ppf(0.01, dfn, dfd), f.ppf(0.99, dfn, dfd), 100)`: `np.linspace`   F     `x` . `f.ppf`   F  (Percent Point Function) ,  0.01 0.99 100   .

#4. `fv = f(dfn, dfd)`: `f(dfn, dfd)`  F   .   F   ,   (PDF)   .

#5. `fig, ax = plt.subplots(1,1)`: Matplotlib  (figure)  (subplot) . `fig`   , `ax`    .

#6. `ax.plot(x, fv.pdf(x), 'g-', lw=1)`: (`ax`) F     . `x`   , `fv.pdf(x)`      PDF  . 'g-'    , `lw=1`   .

#7. `plt.title("F distribution: F(40,8)")`:   .   "F distribution: F(40,8)"  .

#  F         .   F  ,    dfn=40  dfd=8  F     .
```

```python
from scipy.stats import f, chi2
df = 20
x = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.99, df), 100)
c_values = chi2.rvs(df, size=1000)

fig, ax = plt.subplots(1,1)

rv = chi2(df)
ax.plot(x, rv.pdf(x), 'r-', lw=2)
ax.hist(c_values, density=True, bins=100, histtype='stepfilled', alpha =0.5)
ax.set_xlim([x[0], x[-1]])
plt.title("Chi-saquare distribution: Chi2(20)")
plt.show()
```
