```python
import random
import matplotlib.pyplot as plt

play = 1000
head = 0
total = 0
final =[]

for i in range(1, play + 1):

    result = random.choice([0, 1])
    head += result
    total += 1
    
    final1 = head / total
    final.append(final1)

plt.plot(range(1, play + 1), final, color='b')
plt.axhline(y=final[-1], color='r' )

final_value = round(final[-1], 3)
plt.title("Probability is {:.3f}" .format(final_value))
plt.show()
```
