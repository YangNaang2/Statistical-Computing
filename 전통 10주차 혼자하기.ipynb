```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv("boston_1978.csv")
data1.head()
```

```python
sns.pairplot(data=data1, vars = ['CRIM', 'NOX','AGE','DIS'], kind = 'scatter', dropna = True)
```

```python
g = sns.pairplot(data=data1, hue = 'TOWN', vars = ['CRIM', 'NOX', 'AGE','DIS'], dropna = True)
g.map_upper(sns.scatterplot)
g.map_lower(sns.scatterplot)
g.map_diag(sns.histplot)

plt.savefig('hueplot.png')
```

```python
data1.corr(method = 'pearson', numeric_only=True)
```

```python
from scipy.stats import pearsonr

data1_clean = data1.dropna(axis=0, how='any', inplace = False)
## data1  (NaN)   ,    data1_clean . axis=0       , how='any'         . inplace=False     .

r2, p_value = pearsonr(data1_clean['AGE'], data1_clean['ZN'])
##pearsonr   'AGE' 'ZN'     (r2) p-(p_value) 
print('The correlation coefficient is', np.round(r2, 4))
print('p value is,',np.round(p_value,4))
```

```python
import statsmodels.api as sm

added_x = sm.add_constant(data1_clean[['AGE', 'RAD']])
##    .AGE, RAD, .
##add_constant   () 
ols_fitted = sm.OLS(data1_clean['TAX'], added_x).fit()
##sm.OLS    .  = TAX // = added_x
##fit()  . ->       .   .
```

```python
reg_coef = ols_fitted.params
##       .
##params ->     
print('--------regression coefficients ----------')
print(reg_coef)
```

```python
print(ols_fitted.summary())
```

```python
##      ,   
predictive_values = ols_fitted.get_prediction(added_x)
##added_X      .
results = predictive_values.summary_frame(alpha=0.05).round(4)
##  summary_frame  .       
##alpha = 0.05 -> 95% 
results.head()
```

```python
sns.lmplot(x='AGE', y='TAX', data= data1_clean, ci=95)
## lmplot  . age tax      .
##ci=confidence interval . 
```

```python
sns.lmplot(x='RAD', y= 'TAX', data= data1_clean, ci=95)
```
