```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv("boston_1978.csv")
data1.head()
```

```python
sns.pairplot(data=data1, vars = ['CRIM', 'NOX','AGE','DIS'], kind = 'scatter', dropna = True)
```

```python
data1.corr(method = 'pearson', numeric_only=True)
```

```python
from scipy.stats import pearsonr

data1_clean = data1.dropna(axis=0, how='any', inplace = False)
##데이터프레임 data1에서 누락된 값(NaN)이 있는 행을 제거하고, 결과를 새로운 데이터프레임인 data1_clean에 할당합니다. axis=0은 행을 기준으로 누락된 값을 제거하라는 것을 의미하며, how='any'는 행의 어떤 값이라도 누락되면 해당 행을 제거하라는 것을 나타냅니다. inplace=False는 원본 데이터프레임을 변경하지 말라는 의미입니다.

r2, p_value = pearsonr(data1_clean['AGE'], data1_clean['ZN'])
##pearsonr 함수를 사용하여 'AGE'와 'ZN' 변수 간의 피어슨 상관 계수(r2)와 p-값(p_value)을 계산
print('The correlation coefficient is', np.round(r2, 4))
print('p value is,',np.round(p_value,4))
```

```python
import statsmodels.api as sm

added_x = sm.add_constant(data1_clean[['AGE', 'RAD']])
##독립 변수로 사용할 데이터프레임 생성.AGE, RAD,변수 사용.
##add_constant를 사용하여 상수 항(절편)을 추가
ols_fitted = sm.OLS(data1_clean['TAX'], added_x).fit()
##sm.OLS 함수 통해서 선형회귀모델 생성. 종속변수 = TAX //독립변수 = added_x
##fit()통해서 모델을 피팅. ->독립 변수와 종속 변수 사이의 선형 관계 평가. 모델 계수 추정.
```

```python
reg_coef = ols_fitted.params
##선형 회귀 모델에 의해 추정된 모델 계수 나타냄.
##params ->선형 회귀 모델의 계수 포함 속성
print('--------regression coefficients ----------')
print(reg_coef)
```

```python
print(ols_fitted.summary())
```

```python
## 통계모델에서 예측값 및 예측 구간 얻고, 해당 예측값 요약
predictive_values = ols_fitted.get_prediction(added_x)
##added_X에 대한 예측값 및 예측 구간 얻는다.
results = predictive_values.summary_frame(alpha=0.05).round(4)
##객체에 대해 summary_frame 메서드 호출.예측 결과의 요약 정보를 데이터 프레임 형태로 반환
##alpha = 0.05 -> 95% 신뢰수준
results.head()
```
